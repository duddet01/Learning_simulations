```python
import os
import shutil

root = "/content/test1_data"

# 1. Move all files from subdirectories to root
for dirpath, dirnames, filenames in os.walk(root):
    for filename in filenames:
        src = os.path.join(dirpath, filename)
        dst = os.path.join(root, filename)

        if src != dst:
            # If file already exists, overwrite
            if os.path.exists(dst):
                os.remove(dst)
            shutil.move(src, dst)

# 2. Remove all empty subdirectories (bottom-up)
for dirpath, dirnames, filenames in os.walk(root, topdown=False):
    if dirpath != root and not os.listdir(dirpath):
        os.rmdir(dirpath)

print("Done. Final contents:")
print(os.listdir(root))
os.chdir('/content/test1_data')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# load data
T1, av_M1, av_E1, cv1, chi1, av_M11, binder1 = np.loadtxt("ising26.dat", unpack=True)
T2, av_M2, av_E2, cv2, chi2, av_M12, binder2 = np.loadtxt("ising26_1.dat", unpack=True)

# collate datasets
T = np.concatenate((T1, T2))
av_M = np.concatenate((av_M1, av_M2))
av_E = np.concatenate((av_E1, av_E2))
cv = np.concatenate((cv1, cv2))
chi = np.concatenate((chi1, chi2))
av_M1 = np.concatenate((av_M11, av_M12))
binder = np.concatenate((binder1, binder2))

# sort by temperature (important for clean lines)
idx = np.argsort(T)

T = T[idx]
av_M = av_M[idx]
av_E = av_E[idx]
cv = cv[idx]
chi = chi[idx]
av_M1 = av_M1[idx]
binder = binder[idx]

fig, axs = plt.subplots(2,3, figsize=(14,8))

# Absolute Magnetization
axs[0,0].scatter(T, av_M)
axs[0,0].plot(T, av_M)
axs[0,0].set_xlabel("Temperature (kB T)")
axs[0,0].set_ylabel("⟨|M|⟩")
axs[0,0].set_title("Absolute Magnetization")
axs[0,0].grid(True)

# Energy
axs[0,1].scatter(T, av_E)
axs[0,1].plot(T, av_E)
axs[0,1].set_xlabel("Temperature (kB T)")
axs[0,1].set_ylabel("⟨E⟩")
axs[0,1].set_title("Energy")
axs[0,1].grid(True)

# Specific Heat
axs[0,2].scatter(T, cv)
axs[0,2].plot(T, cv)
axs[0,2].set_xlabel("Temperature (kB T)")
axs[0,2].set_ylabel("$C_v$")
axs[0,2].set_title("Specific Heat")
axs[0,2].grid(True)

# Susceptibility
axs[1,0].scatter(T, chi)
axs[1,0].plot(T, chi)
axs[1,0].set_xlabel("Temperature (kB T)")
axs[1,0].set_ylabel("χ")
axs[1,0].set_title("Magnetic Susceptibility")
axs[1,0].grid(True)

# Magnetization
axs[1,1].scatter(T, av_M1)
axs[1,1].plot(T, av_M1)
axs[1,1].set_xlabel("Temperature (kB T)")
axs[1,1].set_ylabel("⟨M⟩")
axs[1,1].set_title("Magnetization")
axs[1,1].grid(True)

# Binder Cumulant
axs[1,2].scatter(T, binder)
axs[1,2].plot(T, binder)
axs[1,2].set_xlabel("Temperature (kB T)")
axs[1,2].set_ylabel("U")
axs[1,2].set_title("Binder Cumulant")
axs[1,2].grid(True)

fig.suptitle("Thermodynamic Quantities vs Temperature (Triangular Ising Model) for L=26", fontsize=16)

plt.tight_layout()
plt.show()
```

```python
#Dataset with L=30

import numpy as np
import matplotlib.pyplot as plt

# -------- load first dataset --------
T_a, Mabs_a, E_a, Cv_a, Chi_a, Mavg_a, Binder_a = np.loadtxt("ising30.dat", unpack=True)

# -------- load second dataset --------
T_b, Mabs_b, E_b, Cv_b, Chi_b, Mavg_b, Binder_b = np.loadtxt("ising30_1.dat", unpack=True)

# -------- collate datasets --------
T_all = np.concatenate((T_a, T_b))
Mabs_all = np.concatenate((Mabs_a, Mabs_b))
E_all = np.concatenate((E_a, E_b))
Cv_all = np.concatenate((Cv_a, Cv_b))
Chi_all = np.concatenate((Chi_a, Chi_b))
Mavg_all = np.concatenate((Mavg_a, Mavg_b))
Binder_all = np.concatenate((Binder_a, Binder_b))

# -------- sort by temperature --------
order = np.argsort(T_all)

T_sorted = T_all[order]
Mabs_sorted = Mabs_all[order]
E_sorted = E_all[order]
Cv_sorted = Cv_all[order]
Chi_sorted = Chi_all[order]
Mavg_sorted = Mavg_all[order]
Binder_sorted = Binder_all[order]

# -------- subplot figure --------
fig, ax = plt.subplots(2,3, figsize=(14,8))

# Absolute Magnetization
ax[0,0].scatter(T_sorted, Mabs_sorted)
ax[0,0].plot(T_sorted, Mabs_sorted)
ax[0,0].set_title("Absolute Magnetization")
ax[0,0].set_xlabel("Temperature (kB T)")
ax[0,0].set_ylabel("<|M|>")
ax[0,0].grid()

# Energy
ax[0,1].scatter(T_sorted, E_sorted)
ax[0,1].plot(T_sorted, E_sorted)
ax[0,1].set_title("Energy per Spin")
ax[0,1].set_xlabel("Temperature (kB T)")
ax[0,1].set_ylabel("<E>")
ax[0,1].grid()

# Specific Heat
ax[0,2].scatter(T_sorted, Cv_sorted)
ax[0,2].plot(T_sorted, Cv_sorted)
ax[0,2].set_title("Specific Heat")
ax[0,2].set_xlabel("Temperature (kB T)")
ax[0,2].set_ylabel("Cv")
ax[0,2].grid()

# Susceptibility
ax[1,0].scatter(T_sorted, Chi_sorted)
ax[1,0].plot(T_sorted, Chi_sorted)
ax[1,0].set_title("Magnetic Susceptibility")
ax[1,0].set_xlabel("Temperature (kB T)")
ax[1,0].set_ylabel("χ")
ax[1,0].grid()

# Magnetization
ax[1,1].scatter(T_sorted, Mavg_sorted)
ax[1,1].plot(T_sorted, Mavg_sorted)
ax[1,1].set_title("Magnetization")
ax[1,1].set_xlabel("Temperature (kB T)")
ax[1,1].set_ylabel("<M>")
ax[1,1].grid()

# Binder Cumulant
ax[1,2].scatter(T_sorted, Binder_sorted)
ax[1,2].plot(T_sorted, Binder_sorted)
ax[1,2].set_title("Binder Cumulant")
ax[1,2].set_xlabel("Temperature (kB T)")
ax[1,2].set_ylabel("U")
ax[1,2].grid()

fig.suptitle("Thermodynamic Quantities vs Temperature (L = 30)", fontsize=16)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- helper function ----------
def load_and_merge(file1, file2):
    T1, M1, E1, Cv1, Chi1, Mavg1, Binder1 = np.loadtxt(file1, unpack=True)
    T2, M2, E2, Cv2, Chi2, Mavg2, Binder2 = np.loadtxt(file2, unpack=True)

    T = np.concatenate((T1, T2))
    Cv = np.concatenate((Cv1, Cv2))
    Chi = np.concatenate((Chi1, Chi2))
    Binder = np.concatenate((Binder1, Binder2))

    idx = np.argsort(T)

    return T[idx], Cv[idx], Chi[idx], Binder[idx]


# ---------- load merged datasets ----------
T26, Cv26, Chi26, Binder26 = load_and_merge("ising26.dat","ising26_1.dat")
T30, Cv30, Chi30, Binder30 = load_and_merge("ising30.dat","ising30_1.dat")
T34, Cv34, Chi34, Binder34 = load_and_merge("ising34.dat","ising34_1.dat")


# ---------- find peaks ----------
Cv26_i = np.argmax(Cv26)
Cv30_i = np.argmax(Cv30)
Cv34_i = np.argmax(Cv34)

Chi26_i = np.argmax(Chi26)
Chi30_i = np.argmax(Chi30)
Chi34_i = np.argmax(Chi34)


# ---------- Binder intersection ----------
diff = np.abs(Binder30 - Binder34)
cross_i = np.argmin(diff)

Tc = T30[cross_i]
Uc = Binder30[cross_i]


# =====================================================
# Specific Heat
# =====================================================

plt.figure(figsize=(7,5))

plt.plot(T26, Cv26, marker='o', label="L = 26")
plt.plot(T30, Cv30, marker='s', label="L = 30")
plt.plot(T34, Cv34, marker='^', label="L = 34")

plt.scatter(T26[Cv26_i], Cv26[Cv26_i])
plt.scatter(T30[Cv30_i], Cv30[Cv30_i])
plt.scatter(T34[Cv34_i], Cv34[Cv34_i])

plt.annotate(f"({T26[Cv26_i]:.2f},{Cv26[Cv26_i]:.2f})",
             (T26[Cv26_i], Cv26[Cv26_i]), xytext=(10,10),
             textcoords='offset points', arrowprops=dict(arrowstyle="->"))

plt.annotate(f"({T30[Cv30_i]:.2f},{Cv30[Cv30_i]:.2f})",
             (T30[Cv30_i], Cv30[Cv30_i]), xytext=(10,10),
             textcoords='offset points', arrowprops=dict(arrowstyle="->"))

plt.annotate(f"({T34[Cv34_i]:.2f},{Cv34[Cv34_i]:.2f})",
             (T34[Cv34_i], Cv34[Cv34_i]), xytext=(10,10),
             textcoords='offset points', arrowprops=dict(arrowstyle="->"))

plt.xlabel("Temperature (kB T)")
plt.ylabel("Specific Heat $C_v$")
plt.title("Specific Heat vs Temperature",pad=15)
plt.legend()
plt.grid()
plt.show()


# =====================================================
# Susceptibility
# =====================================================

plt.figure(figsize=(7,5))

plt.plot(T26, Chi26, marker='o', label="L = 26")
plt.plot(T30, Chi30, marker='s', label="L = 30")
plt.plot(T34, Chi34, marker='^', label="L = 34")

plt.scatter(T26[Chi26_i], Chi26[Chi26_i])
plt.scatter(T30[Chi30_i], Chi30[Chi30_i])
plt.scatter(T34[Chi34_i], Chi34[Chi34_i])

plt.annotate(f"({T26[Chi26_i]:.2f},{Chi26[Chi26_i]:.2f})",
             (T26[Chi26_i], Chi26[Chi26_i]), xytext=(10,10),
             textcoords='offset points', arrowprops=dict(arrowstyle="->"))

plt.annotate(f"({T30[Chi30_i]:.2f},{Chi30[Chi30_i]:.2f})",
             (T30[Chi30_i], Chi30[Chi30_i]), xytext=(10,10),
             textcoords='offset points', arrowprops=dict(arrowstyle="->"))

plt.annotate(f"({T34[Chi34_i]:.2f},{Chi34[Chi34_i]:.2f})",
             (T34[Chi34_i], Chi34[Chi34_i]), xytext=(10,10),
             textcoords='offset points', arrowprops=dict(arrowstyle="->"))

plt.xlabel("Temperature (kB T)")
plt.ylabel("Magnetic Susceptibility χ")
plt.title("Susceptibility vs Temperature",pad=15)
plt.legend()
plt.grid()
plt.show()


# =====================================================
# Binder Cumulant
# =====================================================

plt.figure(figsize=(7,5))

plt.plot(T26, Binder26, marker='o', label="L = 26")
plt.plot(T30, Binder30, marker='s', label="L = 30")
plt.plot(T34, Binder34, marker='^', label="L = 34")

plt.scatter(Tc, Uc)

plt.annotate(f"({Tc:.2f},{Uc:.2f})",
             (Tc,Uc), xytext=(8,10),
             textcoords='offset points', arrowprops=dict(arrowstyle="->"))

plt.xlabel("Temperature (kB T)")
plt.ylabel("Binder Cumulant U")
plt.title("Binder Cumulant vs Temperature" ,pad=15)
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- helper function ----------
def load_and_merge(file1, file2):
    T1, M1, E1, Cv1, Chi1, Mavg1, Binder1 = np.loadtxt(file1, unpack=True)
    T2, M2, E2, Cv2, Chi2, Mavg2, Binder2 = np.loadtxt(file2, unpack=True)

    T = np.concatenate((T1, T2))
    Binder = np.concatenate((Binder1, Binder2))

    idx = np.argsort(T)

    return T[idx], Binder[idx]


# ---------- load datasets ----------
T10, B10 = load_and_merge("ising15.dat","ising15_1.dat")
T26, B26 = load_and_merge("ising26.dat","ising26_1.dat")
T30, B30 = load_and_merge("ising30.dat","ising30_1.dat")
T34, B34 = load_and_merge("ising34.dat","ising34_1.dat")


# ---------- zoom region ----------
def zoom(T,B):
    mask = (T >= 3.0) & (T <= 4.0)
    return T[mask], B[mask]

T10z,B10z = zoom(T10,B10)
T26z,B26z = zoom(T26,B26)
T30z,B30z = zoom(T30,B30)
T34z,B34z = zoom(T34,B34)


# ---------- intersection finder ----------
def find_intersection(T, U1, U2):

    diff = U1 - U2

    for i in range(len(diff)-1):
        if diff[i]*diff[i+1] < 0:  # sign change -> crossing

            # linear interpolation
            T_cross = T[i] - diff[i]*(T[i+1]-T[i])/(diff[i+1]-diff[i])
            U_cross = U1[i] + (U1[i+1]-U1[i])*(T_cross-T[i])/(T[i+1]-T[i])

            return T_cross, U_cross

    return None,None


# intersections
Tc_26_30, Uc_26_30 = find_intersection(T26z,B26z,B30z)
Tc_30_34, Uc_30_34 = find_intersection(T30z,B30z,B34z)


# ---------- plot ----------
plt.figure(figsize=(7,5))

plt.plot(T10z, B10z, marker='d', label="L = 10")
plt.plot(T26z, B26z, marker='o', label="L = 26")
plt.plot(T30z, B30z, marker='s', label="L = 30")
plt.plot(T34z, B34z, marker='^', label="L = 34")

# plot intersection points
if Tc_26_30:
    plt.scatter(Tc_26_30, Uc_26_30, s=120)
    plt.annotate(f"Tc≈{Tc_26_30:.3f}",
                 (Tc_26_30,Uc_26_30),
                 xytext=(15,10),
                 textcoords="offset points",
                 arrowprops=dict(arrowstyle="->"))

if Tc_30_34:
    plt.scatter(Tc_30_34, Uc_30_34, s=120)
    plt.annotate(f"Tc≈{Tc_30_34:.3f}",
                 (Tc_30_34,Uc_30_34),
                 xytext=(15,10),
                 textcoords="offset points",
                 arrowprops=dict(arrowstyle="->"))

plt.xlabel("Temperature (kB T)")
plt.ylabel("Binder Cumulant U")
plt.title("Binder Cumulant vs Temperature (Critical Region)",)
plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- helper function ----------
def load_and_merge(file1, file2):

    T1, M1, E1, Cv1, Chi1, Mavg1, Binder1 = np.loadtxt(file1, unpack=True)
    T2, M2, E2, Cv2, Chi2, Mavg2, Binder2 = np.loadtxt(file2, unpack=True)

    T = np.concatenate((T1,T2))
    Binder = np.concatenate((Binder1,Binder2))

    idx = np.argsort(T)

    return T[idx], Binder[idx]


# ---------- load datasets ----------
T9, B9 = load_and_merge("ising9.dat","ising9_1.dat")
T15, B15 = load_and_merge("ising15.dat","ising15_1.dat")
T30, B30 = load_and_merge("ising30.dat","ising30_1.dat")


# ---------- zoom temperature ----------
def zoom(T,B):

    mask = (T >= 3.0) & (T <= 4.0)

    return T[mask], B[mask]


T9z,B9z = zoom(T9,B9)
T15z,B15z = zoom(T15,B15)
T30z,B30z = zoom(T30,B30)


# ---------- find intersection (L15 vs L30) ----------
diff = np.abs(B15z - B30z)

index = np.argmin(diff)

Tc = T15z[index]

print("Estimated Tc from Binder crossing =", Tc)


# ---------- plot ----------
plt.figure(figsize=(7,5))

plt.plot(T9z, B9z, marker='o', label="L = 9")
plt.plot(T15z, B15z, marker='s', label="L = 15")
plt.plot(T30z, B30z, marker='^', label="L = 30")

# single vertical dashed line
plt.axvline(Tc, linestyle='--', color='red',linewidth=0.9)
plt.text(Tc, 0.2, f"Tc ≈ {Tc:.3f}",
         rotation=0,
         verticalalignment='bottom')

plt.xlabel("Temperature (kB T)")
plt.ylabel("Binder Cumulant U")

plt.title("Binder Cumulant vs Temperature (3 ≤ T ≤ 4)", pad=15)

plt.legend()
plt.grid()

plt.show()
```
