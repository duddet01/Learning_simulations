```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
data = np.loadtxt("band.dat")

k = data[:,0]
E = data[:,1:6]

plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(8,6), dpi=120)

# Marker styles (different for each band)
markers = ['o', 's', '^', 'D', 'x']
colors = plt.cm.viridis(np.linspace(0,1,5))

for i in range(5):
    plt.plot(
        k, E[:,i],
        marker=markers[i],
        markevery=25,     # ← controls spacing of markers (IMPORTANT)
        linewidth=1.8,
        markersize=5,
        color=colors[i],
        label=f'Band {i+1}'
    )

# Brillouin zone boundaries
pi = np.pi
plt.axvline(-pi, linestyle='--', linewidth=1, color='black')
plt.axvline( pi, linestyle='--', linewidth=1, color='black')

# Labels
plt.xlabel(r'$k$', fontsize=14)
plt.ylabel(r'Energy $E(k)$', fontsize=14)
plt.title("Band Structure for first brillouin zone (V=0.0 , npw=11 )", fontsize=16)

# k-axis ticks
plt.xticks([-pi, 0, pi], [r'$-\pi/a$', r'$0$', r'$\pi/a$'])

plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
psi_k0 = np.loadtxt("psi_k0.dat")
psi_kpi = np.loadtxt("psi_kpi.dat")

# Extract columns
x0 = psi_k0[:,0]
psi0 = psi_k0[:,1]

xpi = psi_kpi[:,0]
psi1 = psi_kpi[:,1]
psi2 = psi_kpi[:,2]

# (OPTIONAL) Load energies from band.dat
band = np.loadtxt("band.dat")

# Find k closest to 0 and -pi
k_vals = band[:,0]
idx_k0 = np.argmin(np.abs(k_vals - 0))
idx_kpi = np.argmin(np.abs(k_vals + np.pi))

E_k0 = band[idx_k0,1]     # lowest band at k=0
E1_kpi = band[idx_kpi,1]  # lowest at k=-pi/a
E2_kpi = band[idx_kpi,2]  # second band

# Create subplots
plt.style.use('seaborn-v0_8-whitegrid')
fig, axs = plt.subplots(1, 3, figsize=(15,4), dpi=120)

# -------- Panel 1: k = 0 --------
axs[0].plot(x0, psi0, marker='o', markevery=30, linewidth=1.8)
axs[0].set_title(rf'$k=0,\ E={E_k0:.3f}$')
axs[0].set_xlabel('x')
axs[0].set_ylabel(r'$|\psi(x)|^2$')
axs[0].grid(alpha=0.3)

# -------- Panel 2: k = -pi/a (1st eigenstate) --------
axs[1].plot(xpi, psi1, marker='s', markevery=30, linewidth=1.8, color='orange')
axs[1].set_title(rf'$k=-\pi/a,\ E_1={E1_kpi:.3f}$')
axs[1].set_xlabel('x')
axs[1].grid(alpha=0.3)

# -------- Panel 3: k = -pi/a (2nd eigenstate) --------
axs[2].plot(xpi, psi2, marker='^', markevery=30, linewidth=1.8, color='green')
axs[2].set_title(rf'$k=-\pi/a,\ E_2={E2_kpi:.3f}$')
axs[2].set_xlabel('x')
axs[2].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

files = ["band_3.dat", "band_11.dat", "band_21.dat", "band_61.dat"]
labels = ["npw = 3", "npw = 11", "npw = 21", "npw = 61"]
nbands_list = [3, 5, 5, 5]

plt.style.use('seaborn-v0_8-whitegrid')

fig, axs = plt.subplots(2, 2, figsize=(10,8), dpi=120)
axs = axs.flatten()

for i, f in enumerate(files):

    data = np.genfromtxt(f, invalid_raise=False)
    data = data[~np.isnan(data).any(axis=1)]

    k = data[:,0]
    E = data[:,1:6]

    for band in range(nbands_list[i]):
        axs[i].plot(
            k, E[:,band],
            marker='o',
            markevery=40,
            linewidth=1.5,
            markersize=4
        )

    axs[i].set_title(labels[i])
    axs[i].set_xlabel("k")
    axs[i].set_ylabel("E")
    axs[i].grid(alpha=0.3)

pi = np.pi
for ax in axs:
    ax.set_xticks([-pi, 0, pi])
    ax.set_xticklabels([r'$-\pi/a$', r'$0$', r'$\pi/a$'])

fig.suptitle("a=1.0 ,v=0.0 , nk =401")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

# Files and corresponding npw
files = ["band_44_3.dat", "band_44_11.dat", "band_44_21.dat", "band_44_61.dat"]
npw_vals = [3, 11, 21, 61]

E1_list = []
E2_list = []

# 🔥 choose non-zero k to avoid trivial flat result
k_target = 1.0

for f in files:
    data = np.genfromtxt(f, invalid_raise=False)

    # keep only valid rows (based on k column)
    data = data[~np.isnan(data[:,0])]

    k = data[:,0]

    # find closest k
    idx = np.argmin(np.abs(k - k_target))

    E1_list.append(data[idx,1])
    E2_list.append(data[idx,2])

E1_list = np.array(E1_list)
E2_list = np.array(E2_list)

# 🔷 Plot
plt.figure(figsize=(7,5), dpi=120)

plt.plot(npw_vals, E1_list, marker='o', linewidth=2, label='E1 (lowest band)')
plt.ylim(0.494,0.496)

plt.xlabel("Number of Plane Waves (npw)")
plt.ylabel("Energy")
plt.title(rf"Convergence of Energy at k ≈ {k_target}")

plt.grid(alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

# Load data (npw = 11 → ng = 5)
data = np.genfromtxt("band_V_11.dat", invalid_raise=False)

# keep valid rows
data = data[~np.isnan(data[:,0])]

k = data[:,0]
E1 = data[:,1]
E2 = data[:,2]

# find Brillouin zone boundary
pi = np.pi
k_boundary = pi / 1.0   # since a = 1

idx = np.argmin(np.abs(k - k_boundary))

E1_gap = E1[idx]
E2_gap = E2[idx]
gap = E2_gap - E1_gap

# 🔷 Plot
plt.figure(figsize=(7,5), dpi=120)

# plot only first 2 bands
plt.plot(k, E1,
         linewidth=2, linestyle='--' , label='E1')

plt.plot(k, E2,
         linewidth=2, linestyle='--', label='E2')




# vertical boundary lines
plt.axvline(k_boundary, linestyle='--', color='black', alpha=0.4)
plt.axvline(-k_boundary, linestyle='--', color='black', alpha=0.4)

# labels
plt.xlabel("k")
plt.ylabel("Energy")
plt.title(r"Band Gap at Brillouin Zone Boundary (V0 = 0.5, npw = 11)")

plt.xticks([-pi, 0, pi], [r'$-\pi/a$', r'$0$', r'$\pi/a$'])

plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# print gap
print(f"Band gap ≈ {gap:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
data = np.loadtxt("gap_vs_V0.dat")

V0 = data[:,0]
gap = data[:,1]
ratio = data[:,2]

plt.style.use('seaborn-v0_8-whitegrid')

fig, axs = plt.subplots(1, 2, figsize=(12,4), dpi=120)

# -------------------------------
# Plot 1: Gap vs V0
# -------------------------------
axs[0].plot(V0, gap, marker='o', linewidth=2, label='Numerical gap')
axs[0].plot(V0, V0, 'r--', linewidth=2, label='Perturbation (ΔE = V0)')

axs[0].set_xlabel("V0")
axs[0].set_ylabel("Band Gap ΔE")
axs[0].set_title("Gap vs V0")
axs[0].legend()
axs[0].grid(alpha=0.3)

# -------------------------------
# Plot 2: Ratio (gap/V0)
# -------------------------------
axs[1].plot(V0, ratio, marker='s', linewidth=2)

# reference line
axs[1].axhline(1.0, linestyle='--', color='red', label='Perturbation valid')

# mark breakdown region (when deviation >10%)
threshold = 0.9
axs[1].axhline(threshold, linestyle='--', color='black', alpha=0.6)

axs[1].set_xlabel("V0")
axs[1].set_ylabel("ΔE / V0")
axs[1].set_title("Validity of Perturbation Theory")
axs[1].legend()
axs[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
