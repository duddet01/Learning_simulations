```python

import gdown

folder_id = "https://drive.google.com/drive/folders/1XEf6H1G5t9kXo70bc0hWNxkoXRttqy8K?usp=sharing"
gdown.download_folder(id=folder_id, output="test_data", quiet=False)

import os
import shutil

root = "/content/test2_data"

# 1. Move all files from subdirectories to root
for dirpath, dirnames, filenames in os.walk(root):
    for filename in filenames:
        src = os.path.join(dirpath, filename)
        dst = os.path.join(root, filename)

        if src != dst:
            # If file already exists, overwrite
            if os.path.exists(dst):
                os.remove(dst)
            shutil.move(src, dst)

# 2. Remove all empty subdirectories (bottom-up)
for dirpath, dirnames, filenames in os.walk(root, topdown=False):
    if dirpath != root and not os.listdir(dirpath):
        os.rmdir(dirpath)

print("Done. Final contents:")
print(os.listdir(root))
os.chdir('/content/test_data')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

plt.style.use('seaborn-v0_8-whitegrid')

# parameters
V0_list = [0.5, 8.5]
a_list = [1, 4, 7, 10]

fig, axs = plt.subplots(2, 4, figsize=(16,6), dpi=120)
axs = axs.flatten()

plot_id = 0

for V0 in V0_list:
    for a in a_list:

        # load file
        fname = f"band_V0_{int(V0)}_a_{int(a)}.dat"
        data = np.genfromtxt(fname, invalid_raise=False)
        data = data[~np.isnan(data[:,0])]

        k = data[:,0]
        E1 = data[:,1]
        E2 = data[:,2]

        # boundary
        k_boundary = np.pi / a
        idx = np.argmin(np.abs(k - k_boundary))

        gap = E2[idx] - E1[idx]

        ax = axs[plot_id]

        # plot bands
        ax.plot(k, E1, linewidth=2, marker='o',
                markevery=np.arange(5,len(k)-5,40), label='E1')

        ax.plot(k, E2, linewidth=2, marker='s',
                markevery=np.arange(5,len(k)-5,40), label='E2')

        # mark gap
        ax.scatter([k_boundary], [E1[idx]], color='red', zorder=5)
        ax.scatter([k_boundary], [E2[idx]], color='red', zorder=5)
        ax.plot([k_boundary, k_boundary],
                [E1[idx], E2[idx]], 'r--', linewidth=1.5)

        # vertical lines
        ax.axvline(k_boundary, linestyle='--', color='black', alpha=0.3)
        ax.axvline(-k_boundary, linestyle='--', color='black', alpha=0.3)

        # labels
        ax.set_title(f"V₀={V0}, a={a}, gap≈{gap:.2f}")
        ax.set_xlabel("k")
        ax.set_ylabel("E")

        ax.set_xticks([-np.pi/a, 0, np.pi/a])
        ax.set_xticklabels([r'$-\pi/a$', r'$0$', r'$\pi/a$'])

        ax.grid(alpha=0.3)

        plot_id += 1

# super title
fig.suptitle("Band Structure & Band Gaps vs Lattice Spacing (nk=401, npw=11)",
             fontsize=16)

plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```

```python
fig, axs = plt.subplots(2, 4, figsize=(16,6), dpi=120)
axs = axs.flatten()

plot_id = 0

for V0 in V0_list:
    for a in a_list:

        fname = f"psi_kpi_V0_{int(V0)}_a_{int(a)}.dat"
        data = np.loadtxt(fname)

        x = data[:,0]
        psi1 = data[:,1]
        psi2 = data[:,2]

        ax = axs[plot_id]

        ax.plot(x, psi1, label='E1', linewidth=2)
        ax.plot(x, psi2, label='E2', linewidth=2)

        ax.set_title(f"V₀={V0}, a={a}")
        ax.set_xlabel("x")
        ax.set_ylabel(r"$|\psi(x)|^2$")
        ax.legend()

        ax.grid(alpha=0.3)

        plot_id += 1

fig.suptitle("Charge Density at k = π/a (Standing Waves)", fontsize=16)

plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```
